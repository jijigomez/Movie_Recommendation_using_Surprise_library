```python
# Create user-item matrix
user_item_matrix = book_user_rating.pivot_table(index='User-ID', columns='ISBN', values='Book-Rating', fill_value=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix)
```

```python
user_similarity
```

```python
def recommend_books(user_id, user_similarity_matrix, user_item_matrix, k=5):
    # Get similarity scores for the given user
    user_similarity_scores = user_similarity_matrix[user_id]

    # Find users most similar to the given user
    similar_users = user_similarity_scores.argsort()[::-1][1:k+1]  # Exclude itself, select top k similar users

    # Find books that similar users have rated highly and the given user has not rated
    recommended_books = {}
    for similar_user in similar_users:
        similar_user_ratings = user_item_matrix.iloc[similar_user]
        for isbn, rating in similar_user_ratings.items():  # Changed iteritems() to items()
            if rating > 0 and user_item_matrix.loc[user_id, isbn] == 0:
                if isbn in recommended_books:
                    recommended_books[isbn] += user_similarity_scores[similar_user] * rating
                else:
                    recommended_books[isbn] = user_similarity_scores[similar_user] * rating

    # Sort recommended books by weighted rating
    recommended_books = sorted(recommended_books.items(), key=lambda x: x[1], reverse=True)

    return recommended_books[:k]  # Return top k recommended books
```

```python
user_id = 75
recommended_books = recommend_books(user_id, user_similarity, user_item_matrix, k=5)

# Print the recommendations
print(f"Top 5 Recommendations for User {user_id}:")
for i, (isbn, score) in enumerate(recommended_books, 1):
    # Find the book title corresponding to the ISBN
    book_title = df.loc[df['ISBN'] == isbn, 'Book-Title'].iloc[0]
    print(f"{i}. ISBN: {isbn}, Title: {book_title}, Score: {score}")
```
